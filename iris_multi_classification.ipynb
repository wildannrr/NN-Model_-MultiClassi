```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('https://drive.google.com/uc?id=1roJ83AbgzDcvRr0Gwud0BmdUQx-oSG-w')
df.head()
```

```python
df = df.drop(columns='Id')
```

```python
# one hot encoding
category = pd.get_dummies(df.Species, dtype=int)
category
```

```python
# menggabungkan kolom hasil one hot encoding dan membuang kolom spesies karena sudah ada variabel baru dan tidak akan terpakai lagi.

new_df = pd.concat([df, category], axis=1)
new_df = new_df.drop(columns='Species')
new_df
```

```python
# melakukan konversi dataframe tersebut menjadi numpy array dengan fungsi values dari dataframe.
dataset = new_df.values
dataset
```

```python
# menentukan terlebih dahulu variabel dependen dan independen

# Pilih 4 kolom pertama untuk dijadikan sebagai atribut
X = dataset[:,0:4]
# Pilih 3 kolom terakhir sebagai label
y = dataset[:,4:7]
```

```python
# Normalisasi
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
# split data training vs data testing
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
# arsitektur model menggunakan 3 layer

model = Sequential([
                    Dense(64, activation='relu', input_shape=(4,)),
                    Dense(64, activation='relu'),
                    Dense(3, activation='softmax')])
```

```python
# menentukan optimizer dan loss funtion, categorical_crossentropy digunanakan untuk masalah klasifikasi multi class
model.compile(optimizer='Adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train, epochs=100)
```

```python
# akurasi prediksi menggunakan data testing
model.evaluate(X_test, Y_test, batch_size=1)
```
